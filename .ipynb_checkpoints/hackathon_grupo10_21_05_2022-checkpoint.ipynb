# 1. Análise e Tratamento dos Dados


## União das tabelas: juntar as informações presentes nos Datasets de forma **correta**


```python
df = train.merge(features, how = 'left', on = ['Store','Date']).merge(stores, how = 'left', on = 'Store')
```

Removemos os IsHoliday_x da tabela train pois temos nas duas tabelas

```python
df = df.drop('IsHoliday_x', axis = 1)
#df.rename({'IsHoliday_y':'IsHoliday'}, inplace = True)
```

## EDA: Avalie os dados presentes no Dataset final;

```python
df.info()
```

Temos que converter Store e Dept para categoria.

```python
df['Store'] = df.Store.astype('category')
df['Dept'] = df.Dept.astype('category')
df['Type'] = df.Dept.astype('category')
```

```python
df.Date = pd.to_datetime(df['Date'].astype(str), format='%Y-%m-%d')
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
df['Day'] = df.Date.dt.day
df.drop('Date', axis = 1, inplace=True)
```

```python
df.describe()
```

```python
def scatterplots(df, columns, target, hue=None):
    n = len(columns)
    kk = int(n / 3)
    r = 0
    j = 0
    if n % 3 > 0:
        r = 1

    fig, ax = plt.subplots(nrows=kk + r, ncols=3, constrained_layout=True, figsize=(20, 20))

    for i, col in enumerate(columns):
        k = int(i / 3)
        j = i % 3
        sns.scatterplot(x=col, data=df, y=target, hue=hue, ax=ax[k, j])
        for label in ax[k, j].get_xticklabels():
            label.set_rotation(45)
    # removing unused axes
    for i in range(j + 1, 3):
        fig.delaxes(ax[kk + r - 1, i]) 
```

```python
target = 'Weekly_Sales'
scatterplots(df, [col for col in df.select_dtypes(include='number').columns if col != target], target, hue=None) 
```

## Tratamento dos dados

Escalonamento dos dados numéricos

```python
def feat_scale(X):

  # Calcular a média de todas as variáveis
  mu = np.mean(X, axis = 0)

  # Calcular o devio padrão de todas as variáveis
  sigma = np.std(X, axis=0, ddof=1)

  # Fazer os escalonamento das variáveis
  xNorm = (X - mu)/sigma

  return xNorm 
```

```python
dfCat = df_clean.select_dtypes(include = ['category','bool'])
dfNum = df_clean.select_dtypes(include='number')
```

```python
dfNum = dfNum.apply(feat_scale)
```

Transforme as variáveis categóricas;

```python
to_encode = ['Store','Dept','Type']
dfCatDummy = pd.get_dummies(dfCat, columns=to_encode, drop_first=False)
```

Reunindo os datasets novamente

```python
df_full = pd.concat([dfNum, dfCatDummy], axis=1)
```

```python
df_full.head()
```

Preenchemos os NaN das colunas MarkDown com 0 (eram as únicas com nulos)

```python
columns = ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']
for col in columns:
  df_full[col].fillna(0, inplace = True)
```

# 2. Análise Micro

 ## Treine modelos com uma quantidade reduzida de dados

```python
df_subset = df_full.sample(1000)
df_subset.head()
```

```python
X_subset = df_subset.drop('Weekly_Sales', axis = 1)
y_subset = df_subset['Weekly_Sales']
```
